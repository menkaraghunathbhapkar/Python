```python
def random_rgb_tuple():
    color = list(np.random.random(size=3) * 256)
    color = [int(x) for x in color]
    color = tuple(color)
    return color
```

```python
# Python code to illustrate split() function
img=cv2.imread("./BGU_ex/fragment_0002.png")
color=['red','blue','black','yellow','purple']
j=0
with open("./BGU_ex/alignments.txt", "r") as file:
    data = file.readlines()
    
    for line in data:
        final_points_list=[]
        word = line.split()
        
        if(word[0]=='Node'):
            continue
     
        elif((word[0]=='0') and(word[1]=='1')):
            index_line= word.index('line')
            #print("Index of Line is:",index_line)
            points_after_line=word[index_line+1:]
            #print("These are the points afterLine: ",points_after_line)
            l1=[]
            for element_after_line in points_after_line:
                int_element_after_line=int(float(element_after_line))
                #print(int_element_after_line)
                l1.append(int_element_after_line)
                if(len(l1)!=2):
                    continue
                else:
                    final_points_list.append(l1)
                    l1=[]
            print("final points are here",final_points_list)
            #img=cv2.imread("./BGU_ex/fragment_0001.png")
            
            
            pts = np.array(final_points_list)
            plt.scatter(pts[:, 0], pts[:, 1], marker=".", color=color[j], s=200)
            j=j+1

            plt.imshow(img)
            plt.show()


            
```
